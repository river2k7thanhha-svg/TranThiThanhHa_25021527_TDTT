```python
        
# Bài 2: Tính diện tích trồng cây
PI = 3.14  # 
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))
# 1. Diện tích hình chữ nhật
S_cn = a * b
# 2. Bán kính hình tròn (r = min(a, b) / 2)
r = min(a, b) / 2
# 3. Diện tích hình tròn (khu vui chơi)
S_ht = PI * (r ** 2)
# 4. Diện tích trồng cây
S_cay = S_cn - S_ht
# In kết quả với 2 chữ số phần thập phân 
print(f"Diện tích còn để trồng cây là: {S_cay:.2f}")
```

```python
# Bài 5: Chữ cái thường liền trước

    c_hoa = input("Nhập vào một chữ cái hoa (ví dụ: A, B, C, ...): ")
    
    # Kiểm tra: 1 kí tự, phải là chữ cái, và phải là chữ hoa
    if len(c_hoa) != 1 or not c_hoa.isalpha() or not c_hoa.isupper():
        print("Lỗi: Vui lòng nhập *một* chữ cái *hoa* hợp lệ.")
        return

    # Xử lý trường hợp đặc biệt: 'A'
    if c_hoa == 'A': # 
        # Chữ cái thường tương ứng là 'a'. Chữ thường liền trước 'a' là 'z'.
        print("Trường hợp đặc biệt: Chữ cái thường liền trước 'a' là 'z'")
        return
    
    # Chuyển chữ hoa thành chữ thường tương ứng (ví dụ: 'C' -> 'c')
    c_thuong_tuong_ung = c_hoa.lower()
    
    # Lấy giá trị ASCII (ord) của chữ thường tương ứng
    ascii_value = ord(c_thuong_tuong_ung)
    
    # Lấy giá trị ASCII của chữ cái thường liền trước (ví dụ: ord('c') - 1 = ord('b'))
    ascii_prev = ascii_value - 1
    
    # Chuyển lại thành kí tự (chr)
    c_thuong_lien_truoc = chr(ascii_prev)
    
    print(f"Chữ cái thường liền trước của '{c_thuong_tuong_ung}' là: {c_thuong_lien_truoc}")

# bai_tap_5()
```

```python
# Bài 6: Kiểm tra cạnh tam giác và tính diện tích
def bai_tap_6():
    try:
        a = float(input("Nhập cạnh a: "))
        b = float(input("Nhập cạnh b: "))
        c = float(input("Nhập cạnh c: "))
        
        # Điều kiện 1: Các cạnh phải dương
        if a <= 0 or b <= 0 or c <= 0:
            print("Khong phai 3 canh cua tam giac") # 
            return
            
        # Điều kiện 2: Tổng 2 cạnh lớn hơn cạnh còn lại 
        if (a + b > c) and (a + c > b) and (b + c > a):
            # Tính nửa chu vi
            p = (a + b + c) / 2
            
            # Tính diện tích theo công thức Heron
            S = math.sqrt(p * (p - a) * (p - b) * (p - c))
            
            # In diện tích với 1 chữ số phần thập phân 
            print(f"3 cạnh tạo thành tam giác. Diện tích tam giác là: {S:.1f}")
        else:
            print("Khong phai 3 canh cua tam giac") # 
            
    except ValueError:
        print("Lỗi: Đầu vào không phải là số hợp lệ.")

# bai_tap_6()
```

```python
# Bài 7: In ra kí tự thứ 5 và thứ 9 của chuỗi
def bai_tap_7():
    chuoi = input("Nhập vào một chuỗi (kí tự thường, độ dài >= 20): ")
    
    # Kiểm tra độ dài chuỗi
    if len(chuoi) < 20: # 
        print(f"Lỗi: Độ dài chuỗi phải >= 20. Chuỗi bạn nhập có độ dài: {len(chuoi)}")
        return
        
    # Kí tự thứ năm (chỉ mục 4)
    ky_tu_thu_5 = chuoi[4]
    
    # Kí tự thứ chín (chỉ mục 8)
    ky_tu_thu_9 = chuoi[8]
    
    print(f"Kí tự thứ năm của chuỗi là: '{ky_tu_thu_5}'")
    print(f"Kí tự thứ chín của chuỗi là: '{ky_tu_thu_9}'")

# bai_tap_7()
```

```python
# Bài 8: Tính tiền điện
    BAC_GIA = [
        (50, 1984),    # Bậc 1: 0 - 50 kWh
        (50, 2050),    # Bậc 2: 51 - 100 kWh (50 số tiếp theo)
        (100, 2380),   # Bậc 3: 101 - 200 kWh (100 số tiếp theo)
        (100, 2998),   # Bậc 4: 201 - 300 kWh (100 số tiếp theo)
        (100, 3350),   # Bậc 5: 301 - 400 kWh (100 số tiếp theo)
        (float('inf'), 3460) # Bậc 6: 401 trở đi
    ]
    VAT_RATE = 0.08 # 8% VAT [cite: 20]
    # Nhập thông tin [cite: 14]
    ho_ten = input("Nhập họ tên chủ hộ: ")
    try:
        chi_so_truoc = int(input("Nhập chỉ số điện kế tháng trước (kWh): "))
        chi_so_nay = int(input("Nhập chỉ số điện kế tháng này (kWh): "))
    except ValueError:
        print("Lỗi: Chỉ số điện kế phải là số nguyên.")
        return
    # Tính lượng điện tiêu thụ
    so_dien_tieu_thu = chi_so_nay - chi_so_truoc
    # Bắt đầu tính tiền
    tong_tien_chua_vat = 0
    con_lai = so_dien_tieu_thu

    for gioi_han, don_gia in BAC_GIA:
        if con_lai <= 0:
            break
        # Lượng điện tính cho bậc hiện tại
        so_dien_bac_nay = min(con_lai, gioi_han)
        # Cộng dồn tiền điện
        tien_bac_nay = so_dien_bac_nay * don_gia
        tong_tien_chua_vat += tien_bac_nay
        # Cập nhật số điện còn lại
        con_lai -= so_dien_bac_nay
    # Tính tiền VAT
    tien_vat = tong_tien_chua_vat * VAT_RATE
    # Tổng tiền phải trả
    tong_tien_phai_tra = tong_tien_chua_vat + tien_vat
    # Chỉ lấy đến đơn vị đồng (làm tròn xuống/floor) [cite: 20]
    tien_phai_tra_lam_tron = math.floor(tong_tien_phai_tra)
    # In kết quả [cite: 21, 22]
    print(f"\nHo va ten: {ho_ten}")
    print(f"Tiền phải trả là: {tien_phai_tra_lam_tron}")
```
