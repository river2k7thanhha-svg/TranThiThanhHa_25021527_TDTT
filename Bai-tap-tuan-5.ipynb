```python
#W5A1: Viet ham truyen vao 2 so nguyen a,b , tra ve so lon hon
def Max_Of_Two(a,b) -> int:
    if a > b:
        return a
    else:
        return b
a = int(input())
b = int(input())
print(Max_Of_Two(a,b))
```

```python
#W5A2: viet ham bien doi 2 so nguyen a, b 
def swap(a,b) -> tuple(int, int):
    return b, a
a, b = map(int, input().split())
a, b = swap(a,b)
print(a, b)
```

```python
#W5A3: viet ham kiem tra cos phai so nguyen to khong?
def Check_Prime(n) -> bool:
    if n < 2:
        return False
    for i in range(2, math.isqrt(n)+1):
        if n%i == 0:
            return False
    return True
import math
n = int(input())
if Check_Prime(n):
    print(f"{n} la so nguyen to")
else:
    print(f"{n} khong phai la so nguyen to")
```

```python
#W5A4: kiem tra so hoan hao
def Check_perfect_number(n) -> bool:
    sum_divisors = 0
    for i in range(1, n):
        if n%i == 0:
            sum_divisors += i
    return sum_divisors == n
n = int(input())
if Check_perfect_number(n):
    print(f"{n} la so hoan hao")
else:
    print(f"{n} khong phai la so hoan hao")
```

```python
#W5A5: viet ham dua vef motj list so nguyen va mot so nguyen duong k, tra ve vi tri k trong list neu khong co trar ve -1
def vi_tri_k(lst, k) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
        return -1
lst = map(int, input().split())
lst = list(lst)
k = int(input())
print(vi_tri_k(lst, k))
```

```python
#W5A6: viet ham tinh giai thua cuar so nguyen duongw n
def gt(n):
    gt=1
    for i in range(1, n+1):
        gt *= i
    return gt
n=int(input())
gt(n)
```

```python
#W5A7:
def tinh_toan(num1, num2, operat):
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        result = num1 / num2
    else:
        return "Toán tử không hợp lệ"
    return round(result, 2)
n1 = float(input("Nhập số thứ nhất: "))
n2 = float(input("Nhập số thứ hai: "))
op = input("Nhập phép toán (+, -, *, /): ")
print("Kết quả:", tinh_toan(n1, n2, op))
```

```python
#W5A9: tong cac chu so cua so nguyen duong n
def tong_chu_so(n):
    tong=0
    n = abs(n)
    for i in str(n):
        tong += int(i)
    return tong
n = int(input())
print(tong_chu_so(n))
```
