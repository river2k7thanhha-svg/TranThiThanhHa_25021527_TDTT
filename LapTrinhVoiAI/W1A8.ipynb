```python
#W1A8: 
Giải thích đoạn code sau in ra cái gì, và vì sao.
for i in range(3):
    print("Al đang học lần", i + 1)
print("Huấn luyện xong!")
#Giải thích:
1.  Công việc thực hiện: Vòng lặp chạy 3 lần, mỗi lần in một thông báo. Sau khi kết thúc, in ra thông báo cuối cùng.
2.  Nội dung in ra: 4 dòng.
    Al đang học lần 1
    Al đang học lần 2
    Al đang học lần 3
    Huấn luyện xong!
3.  Lý do: Lệnh `print("Al đang học lần", i + 1)` được lặp 3 lần. Lệnh `print("Huấn luyện xong!")` nằm ngoài vòng lặp nên chỉ chạy 1 lần.
```
