```python
# W1A2: 
# Dòng 1: Khai báo và gán chuỗi "Hà Nội" cho biến 'city'.
city = "Hà Nội"
# Dòng 2: Khai báo và gán số nguyên 2025 cho biến 'year'.
year = 2025
# Dòng 3: In ra màn hình một chuỗi kết hợp các chuỗi cố định và giá trị của biến 'city' và 'year'.
# Các giá trị được tự động chuyển thành chuỗi và nối lại bằng dấu cách.
print("Thành phố: ", city, " - Năm: ", year)
# Giải thích kết quả:
# Kết quả in ra là "Thành phố: Hà Nội - Năm: 2025".
# Python tự động thêm dấu cách giữa các đối số khi dùng print() với dấu phẩy.
```
