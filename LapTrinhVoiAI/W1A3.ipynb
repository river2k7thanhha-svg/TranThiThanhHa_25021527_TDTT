```python
# W1A3:
# Dòng 1: Thiết lập giới hạn trên của dãy số là 4.
n = 4
# Dòng 2: Khởi tạo biến tích lũy tổng là 0.
t = 0
# Dòng 3: Bắt đầu vòng lặp. 'i' sẽ chạy từ 1 (bao gồm) đến n+1 (tức là 5, không bao gồm),
# tức là i sẽ lần lượt là 1, 2, 3, 4.
for i in range(1, n + 1):
    # Dòng 4: Cộng giá trị hiện tại của 'i' vào biến 't' (t = t + i).
    t += i
# Dòng 5: In ra giá trị cuối cùng của biến 't'.
print(t)

# Giải thích kết quả:
# 1. Khởi tạo: t = 0
# 2. Vòng lặp 1: i = 1, t = 0 + 1 = 1
# 3. Vòng lặp 2: i = 2, t = 1 + 2 = 3
# 4. Vòng lặp 3: i = 3, t = 3 + 3 = 6
# 5. Vòng lặp 4: i = 4, t = 6 + 4 = 10
# Kết quả in ra là 10.
```
