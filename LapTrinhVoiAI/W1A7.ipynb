```python
# W1A7: 
# Dòng 1: Yêu cầu người dùng nhập số. input() trả về chuỗi.
num_str = input("Nhập số: ")
# Sử dụng try-except để xử lý lỗi nếu người dùng nhập ký tự không phải số.
try:
    # Lỗi 1 được sửa: Ép kiểu chuỗi (num_str) sang số nguyên (int) để thực hiện phép toán số học.
    num = int(num_str)
    # Lỗi 2 được sửa: Dùng toán tử so sánh bằng (==) thay cho toán tử gán (=).
    # Kiểm tra điều kiện: Nếu số chia lấy dư cho 2 bằng 0.
    if num % 2 == 0:
        # Nếu điều kiện đúng, in ra thông báo "Số chẵn".
        print("Số chẵn")
    # Lỗi 3 được sửa: Thêm dấu hai chấm (:) sau từ khóa 'else'.
    else:
        # Nếu điều kiện sai (số lẻ), in ra thông báo "Số lẻ".
        print("Số lẻ")
except ValueError:
    # In ra thông báo lỗi nếu đầu vào không hợp lệ.
    print("Lỗi: Đầu vào không phải là số hợp lệ.")

# Giải thích kết quả:
# Chương trình sẽ yêu cầu người dùng nhập một số.
# - Nếu nhập số chẵn (ví dụ: 10), kết quả in ra là "Số chẵn".
# - Nếu nhập số lẻ (ví dụ: 7), kết quả in ra là "Số lẻ".
# - Nếu nhập chữ (ví dụ: "a"), kết quả in ra là "Lỗi: Đầu vào không phải là số hợp lệ."
```
