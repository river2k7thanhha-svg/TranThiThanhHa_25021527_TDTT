```python
# W1A4:
# Dòng 1: Khai báo một danh sách (list) các số nguyên.
numbers = [1, 2, 3, 4]
# Dòng 2: Bắt đầu vòng lặp. Biến 'x' sẽ lần lượt nhận giá trị của từng phần tử trong danh sách 'numbers'.
for x in numbers:
    # Dòng 3: Kiểm tra điều kiện: Nếu 'x' chia lấy dư cho 2 bằng 0 (x % 2 == 0)
    # thì 'x' là số chẵn.
    if x % 2 == 0:
        # Dòng 4: Nếu là số chẵn, in ra giá trị của 'x' và thông báo.
        print(x, "là số chẵn")
    # Dòng 5: Ngược lại (else): nếu không phải số chẵn, tức là số lẻ.
    else:
        # Dòng 6: In ra giá trị của 'x' và thông báo.
        print(x, "là số lẻ")

# Giải thích kết quả:
# Chương trình sẽ kiểm tra từng số và in ra 4 dòng tương ứng:
# 1 là số lẻ
# 2 là số chẵn
# 3 là số lẻ
# 4 là số chẵn
```
